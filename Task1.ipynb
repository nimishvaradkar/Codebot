Bar chart or Histogram to visualize a distribution of a categorial or continuous variable (Distribution of Age and Gender in a population).

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_85220.csv", skiprows=4)

# Select year and top countries
year = "2022"
df_year = df[["Country Name", year]]
df_year = df_year.dropna()

# Sort and select top 10 most populated countries
top_countries = df_year.sort_values(by=year, ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 6))
plt.bar(top_countries["Country Name"], top_countries[year]/1e9, color='skyblue')
plt.title(f"Top 10 Most Populous Countries in {year}")
plt.ylabel("Population (Billions)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df_year[year]/1e6, bins=30, color='salmon', edgecolor='black')
plt.title(f"Population Distribution Across Countries ({year})")
plt.xlabel("Population (Millions)")
plt.ylabel("Number of Countries")
plt.grid(True)
plt.show()
```

```python
plt.savefig("population_bar_chart.png", dpi=300)
```
